```python
import re 
import pandas as pd
```

```python
f = open('WhatsApp Chat with Ankit.txt', 'r', encoding='utf-8')
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{2}/\d{2}/\d{4},\s\d{2}:\d{2}\s(?:am|pm)?\s?-'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'date': dates})
df['date'] = pd.to_datetime(
	df['date']
	  .str.replace('\u202f', ' ', regex=False)
	  .str.replace(r' -$', '', regex=True),
	format='%d/%m/%Y, %I:%M %p'
)
df.head()
```

```python
df.shape
```

```python
# separate users and messages

users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'(\w+):\s', message)

    if entry[1:]:   # user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
def create_wordcloud(selected_user, df):
    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]
    wc = WordCloud(width=500, height=500, min_font_size=10, background_color='white')
    df_wc = wc.generate(df['message'].str.cat(sep=" "))
    return df_wc
```

```python
df['day_name'] = df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))
```

```python
df['period'] = period
```

```python
df.sample(5)
```

```python
df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0)
```

```python
import seaborn as sns
plt.figure(figsize=(20, 6))
sns.heatmap(df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.xlabel('Period')
plt.ylabel('Day')
plt.title('Weekly Activity Map')
plt.show()
```
